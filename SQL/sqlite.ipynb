````python
import pandas as pd
from sqlite3 import connect
````

````python
# Connect to `chinook` database.
conn = connect('.\databases\chinook.db')
````

OVERVIEW

````python
# Select data in first 5 rows.
query = """
    SELECT * FROM customers LIMIT 5;
"""

pd.read_sql(query, con=conn)
````

SELECT

````python
# Select data with specific countries and email.
query = """
    SELECT
        firstname,
        lastname,
        firstname || ' ' || lastname AS fullname,
        email,
        country
    FROM customers
    WHERE 
        lower(country) IN ('usa', 'brazil', 'united kingdom') AND
        email like '%hotmail%';
"""

pd.read_sql(query, con=conn)
````

````python
# SELECT data WHERE customerid IN range 10 to 15.
query = """
    SELECT
        firstname,
        lastname,
        firstname || ' ' || lastname AS fullname,
        email,
        country
    FROM customers
    WHERE 
        customerid BETWEEN 10 AND 15;
"""

pd.read_sql(query, con=conn)
````

````python
# SELECT data usINg Regex WHERE first letter of country starts with [UB] or lASt letter of firstname ENDs with 'e'.
query = """
    SELECT
        firstname,
        lAStname,
        firstname || ' ' || lAStname AS fullname,
        email,
        country
    FROM customers
    WHERE 
        country glob '[UB]*' OR firstname glob '*e';
"""

pd.read_sql(query, con=conn)
````

````python
# Count the total number of country.
query = """
    SELECT
        count(distinct country) AS country_count
    FROM customers;
"""

pd.read_sql(query, con=conn)
````

IF CLAUSE

````python
# Create a `Region` column FROM `Country` column.
query = """
    SELECT
        country,
        CASE 
            WHEN lower(country) IN ('usa', 'canada') THEN 'America'
            WHEN lower(country) = 'brazil' THEN 'South America'
            WHEN lower(country) IN ('austria', 'belgium', 'denmark') THEN 'Europe'
        ELSE 
            'Other Region' 
        END AS Region
    FROM customers;
"""

pd.read_sql(query, con=conn)
````

JOIN

````python
# Select data where artist name starts with 'C' and track time less then 100000 milliseconds.
query = """
    SELECT
        artists.artistid AS Artist_ID,
        artists.name AS Artist_Name,
        albums.title AS Album_Name,
        tracks.name AS Track_Name,
        tracks.composer AS Track_Composer,
        tracks.milliseconds AS Track_Time
    FROM artists
    LEFT JOIN albums 
        ON artists.artistid = albums.artistid
    LEFT JOIN tracks 
        ON albums.albumid = tracks.albumid
    WHERE 
        artists.name LIKE 'C%' AND milliseconds < 100000
    LIMIT 10;
"""

pd.read_sql(query, con=conn)
````

````python
# Join column using where clause.
query = """
    SELECT 
        * 
    FROM artists, albums, tracks 
    WHERE 
        artists.artistid = albums.artistid AND
        albums.albumid = tracks.albumid 
    LIMIT 5;
"""

pd.read_sql(query, con=conn)
````

AGGREGATE FUNCTION

````python
# Select avg, sum, min, max, count group by genre.
query = """
    SELECT 
        genres.name AS Genre,
        AVG(bytes) AS avg_bytes, 
        SUM(bytes) AS sum_bytes, 
        MIN(bytes) AS min_bytes, 
        MAX(bytes) AS max_bytes, 
        COUNT(bytes) AS count 
    FROM tracks 
    JOIN genres 
        ON tracks.genreid = genres.genreid 
    WHERE 
        Genre <> 'Latin' 
    GROUP BY Genre 
    HAVING 
        count > 100 
    ORDER BY count DESC;
"""

pd.read_sql(query, con=conn)
````

NESTED QUERY

````python
# Select data with customer segmentation where country in 'USA'.
query = """
    SELECT 
        firstname, 
        lastname, 
        company, 
        coalesce(company, 'End Customer') AS Clean_Company,
        CASE 
            WHEN company IS NULL THEN 'B2B' 
        ELSE 
            'B2C' 
        END AS Segment
    FROM (	
        SELECT 
            * 
        FROM customers 
        WHERE 
            lower(country) = 'usa' 
    );
"""

pd.read_sql(query, con=conn)
````

````python
# Count the total number of customer segmentation.
query = """
    SELECT  
        Segment, 
        count(*) AS n 
    FROM ( 
        SELECT 
            firstname, 
            lastname, 
            company, 
            coalesce(company, 'End Customer') AS Clean_Company,
            CASE 
                WHEN company IS NULL THEN 'B2B' 
            ELSE 
                'B2C' 
            END AS Segment
        FROM (	
            SELECT 
                * 
            FROM customers 
            WHERE 
                lower(country) = 'usa' 
        ) 
    ) 
    GROUP BY Segment;
"""

pd.read_sql(query, con=conn)
````

WINDOW FUNCTION

````python
# Create a running number column partitioning by country.
query = """
    SELECT 
        firstname, 
        lastname, 
        country, 
        ROW_NUMBER() OVER(PARTITION BY country) AS Row_Num 
    FROM customers
    LIMIT 10;
"""

pd.read_sql(query, con=conn)
````

````python
# Select data having running number = 1.
query = """
    SELECT 
        * 
    FROM ( 
        SELECT 
            firstname, 
            lastname, 
            country, 
            ROW_NUMBER() OVER(PARTITION BY country) AS Row_Num 
        FROM customers 
    ) 
    WHERE 
        Row_Num = 1
    LIMIT 10;
"""

pd.read_sql(query, con=conn)
````

````python
# Select data having running number = 1.
query = """
    SELECT 
        title, 
        name, 
        milliseconds, 
        RANK() OVER(PARTITION BY title ORDER BY milliseconds DESC) AS Rank 
    FROM tracks 
    JOIN albums 
        ON tracks.albumid = albums.albumid;
"""

pd.read_sql(query, con=conn)

````
